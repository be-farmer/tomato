```python
import pandas as pd
import numpy as np
```

```python
input_data = pd.read_csv('2023_smartFarm_AI_hackathon_dataset.csv')
input_data
```

```python
input_data.columns
```

```python
input_data.describe()
```

```python
X = input_data.drop(columns=['outtrn_cumsum', 'HeatingEnergyUsage_cumsum', 'frmDist'])
Y = input_data[['outtrn_cumsum', 'HeatingEnergyUsage_cumsum']]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

```python
print(train_input.shape, test_input.shape)
print(train_target.shape, test_target.shape)
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# from sklearn.linear_model import LinearRegression
# model = LinearRegression()
# model.fit(train_scaled, train_target)

# 라쏘 회귀 모델 생성
from sklearn.linear_model import Lasso
alpha = 0.01  # 라쏘 회귀의 alpha 값 설정
lasso_model = Lasso(alpha=alpha)

# 모델 학습
lasso_model.fit(train_scaled, train_target)
```

```python
y_pred = lasso_model.predict(test_scaled)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

rmse = sqrt(mean_squared_error(test_target, y_pred))
r2score = r2_score(test_target, y_pred)

print(f'RMSE: {rmse}')
print(f'R2 Score: {r2score}')
```
